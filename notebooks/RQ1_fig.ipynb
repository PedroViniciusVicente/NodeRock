# **QP 1 - Quão bons são os modelos de aprendizado de máquina para a seleção de testes com event races**

# **Modelos Supervisionados**

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

plt.style.use('default')
plt.rcParams['font.size'] = 11
plt.rcParams['font.family'] = 'serif'

models = ["Random Forest", "KNN (K=5)", "SVM", "Naive Bayes", "Decision Tree", "Ensemble"]
TN = [89, 94, 93, 93, 86, 82]
FP = [ 7,  2,  3,  3, 10, 14]
FN = [13, 17, 16, 27, 15,  8]
TP = [19, 15, 16,  5, 17, 24]

accuracy, precision, recall = [], [], []
for tp, tn, fp, fn in zip(TP, TN, FP, FN):
    accuracy.append((tp + tn) / (tp + tn + fp + fn) * 100)
    precision.append(tp / (tp + fp) * 100)
    recall.append(tp / (tp + fn) * 100)

f1_score = []
for i in range(len(precision)):
    denom = precision[i] + recall[i]
    val = 2 * (precision[i] * recall[i]) / denom if denom != 0 else 0
    f1_score.append(val)

fig, ax = plt.subplots(figsize=(14, 3.5), dpi=300)

table_data = np.array([accuracy, precision, recall, f1_score])
rows = ["Acurácia (%)", "Precisão (%)", "Revocação (%)", "F1-Score (%)"]

cell_text = []
for row in table_data:
    formatted_row = []
    for val in row:
        formatted_row.append(f"{val:.2f}".replace('.', ','))
    cell_text.append(formatted_row)

table = plt.table(cellText=cell_text,
                  rowLabels=rows,
                  colLabels=models,
                  cellLoc='center',
                  rowLoc='center',
                  loc='center',
                  bbox=[0, 0, 1, 1])

table.auto_set_font_size(False)
table.set_fontsize(10)

for i in range(len(models)):
    table[(0, i)].set_facecolor('#E6E6FA')
    table[(0, i)].set_text_props(weight='bold')
for i in range(len(rows)):
    table[(i+1, -1)].set_facecolor('#F0F0F0')
    table[(i+1, -1)].set_text_props(weight='bold')

cor_alto = '#90EE90'
cor_medio = '#FFFFE0'
cor_baixo = '#FFB6C1'
cor_padrao = 'white'

for i in range(len(rows)):
    for j in range(len(models)):
        valor = table_data[i, j]
        if valor >= 80:
            table[(i+1, j)].set_facecolor(cor_alto)
        elif valor >= 70:
            table[(i+1, j)].set_facecolor(cor_medio)
        elif valor < 50:
            table[(i+1, j)].set_facecolor(cor_baixo)
        else:
             table[(i+1, j)].set_facecolor(cor_padrao)

legend_patches = [
    mpatches.Patch(color=cor_alto, label='≥ 80% (Excelente)'),
    mpatches.Patch(color=cor_medio, label='70% - 79% (Bom)'),
    mpatches.Patch(color=cor_padrao, label='50% - 69% (Regular)'),
    mpatches.Patch(color=cor_baixo, label='< 50% (Baixo)')
]
legend_patches[2].set_edgecolor('gray')
legend_patches[2].set_linewidth(0.5)

ax.legend(handles=legend_patches,
          loc='lower center',
          bbox_to_anchor=(0.5, 1.02),
          ncol=4,
          frameon=False,
          title="Legenda de Desempenho")

ax.axis('off')
plt.tight_layout()

plt.show()
```
